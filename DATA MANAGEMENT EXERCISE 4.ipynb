```python
from flask import Flask,request
from flask_mysqldb import MySQL
 
app = Flask(__name__)
 
app.config['MYSQL_HOST'] = 'localhost'
app.config['MYSQL_PORT'] = 3309
app.config['MYSQL_USER'] = 'root'
app.config['MYSQL_PASSWORD'] = 'dragonslayer 12345'
app.config['MYSQL_DB'] = 'mybusiness'
 
mysql = MySQL(app)
```

```python
@app.route('/customers', methods=['GET', 'POST'])
def customers():
    if request.method == 'GET': #  get customers by parameters or all customers
        cursor = mysql.connect.cursor()
        query = 'SELECT * FROM customer'
        parameters = request.args
        if parameters == None:
            cursor.execute((query))
        else:
            query += ' WHERE '
            for key, value in parameters.items():
                if type(value) == str:
                    query += key + ' = ' + f"'{value}'" + ' AND '
                else:
                    query += key + ' = ' + value + ' AND '
            query = query[:-5]
            cursor.execute((query))
        rv = cursor.fetchall()
        dictionaries = []
        for i in rv:
            dictionaries.append({'customer_id': i[0], 'customer_name': i[1], 'city': i[2], 'grade': i[3], 'salesman_id': i[4]})
        cursor.close()
        return dictionaries
        
    elif request.method == 'POST': # add new customer
        cursor = mysql.connection.cursor()
        parameters = request.args
        customer_id = parameters['customer_id']
        customer_name = parameters['customer_name']
        city = parameters['city']
        grade = parameters['grade']
        salesman_id = parameters['salesman_id']
        cursor.execute('INSERT INTO customer (customer_id, cust_name, city, grade, salesman_id) VALUES (%s, %s, %s, %s, %s)', (customer_id, customer_name, city, grade, salesman_id))
        mysql.connection.commit()
        dictionaries = [[{"result" : "added new customer"}]]
        return dictionaries


app.run(host='localhost', port=5000)
```

```python
@app.route('/customers/<customer_id>', methods=['GET', 'PUT', 'DELETE'])
def customer(customer_id):
    if request.method == 'GET': #  get customer by id
        cursor = mysql.connection.cursor()
        cursor.execute(''' SELECT * FROM customer WHERE customer_id = %s''', (customer_id,))
        rv = cursor.fetchall()
        dictionaries = []
        for i in rv:
            dictionaries.append({'customer_id': i[0], 'customer_name': i[1], 'city': i[2], 'grade': i[3], 'salesman_id': i[4]})
        cursor.close()
        return dictionaries

    elif request.method == 'PUT': # update customer by id
        cursor = mysql.connection.cursor()
        parameters = request.args
        query = "UPDATE customer SET "
        for key, value in parameters.items():
            if type(value) == str:
                query += key + ' = ' + f"'{value}'" + ' AND '
            else:
                query += key + ' = ' + value + ' AND '
        query = query[:-5]
        query += f" WHERE customer_id = {customer_id}"
        cursor.execute(query)
        mysql.connection.commit()
        cursor.close()
        dictionaries = [[{"result" : "updated"}]]
        return dictionaries

    elif request.method == 'DELETE': # delete customer by id
        cursor = mysql.connection.cursor()
        cursor.execute('DELETE FROM customer WHERE customer_id = %s', (customer_id,))
        mysql.connection.commit()
        cursor.close()
        dictionaries = [[{"result" : "deleted"}]]
        return dictionaries

app.run(host='localhost', port=5000)
```
